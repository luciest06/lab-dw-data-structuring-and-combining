## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
import numpy as np

ins_df=pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
ins_df2=pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
ins_df3=pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')

print("ins_df shape:",ins_df.shape)
print("ins_df columns:",ins_df.columns)
print(" ")

print("ins_df2 shape:",ins_df2.shape)
print("ins_df2 columns:",ins_df2.columns)

print(" ")
print("ins_df3 shape:",ins_df3.shape)
print("ins_df3 columns:",ins_df3.columns)

#It seems the first and second .csv files have the same columns but in different order.
#Let's realign columns before concatenating both dataframes.
ins_df2 = ins_df2.reindex(columns=ins_df.columns)
ins_df12 = pd.concat([ins_df,ins_df2],axis=0)

#Column names of the third .csv file have a different format.
#Let's clean column names, and then re-order them before concatenating both dataframes

ins_df12.columns=[x.lower() for x in ins_df12.columns]
ins_df12.rename(columns={"st":"state"},inplace=True)

ins_df3.columns=[x.lower() for x in ins_df3.columns]

print("ins_df12 columns:",ins_df12.shape)
print("ins_df12 columns:",ins_df12.columns)

ins_df3 = ins_df3.reindex(columns=ins_df12.columns)

ins_df123 = pd.concat([ins_df12,ins_df3],axis=0)
print(" ")
print("ins_df123 shape:",ins_df123.shape)
print("ins_df123 columns:",ins_df123.columns)

ins_df123.reset_index(drop=True)
```
